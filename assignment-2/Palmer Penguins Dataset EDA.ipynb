```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("penguins_size.csv")
df=df.dropna(axis=0)
df.head()
```

```python
df.dtypes
```

```python
df["body_mass_kg"] = (df["body_mass_g"]/1000)
df["flipper_length_m"] = (df["flipper_length_mm"]/1000)
```

```python
df = df.drop(['flipper_length_mm','body_mass_g'], axis=1)
```

```python
df.head()
```

```python
df.mean()
```

```python
plt.figure(figsize = (8,6))
sns.countplot(df["species"])
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.countplot(df['island'])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(df['sex'])
plt.show()
```
