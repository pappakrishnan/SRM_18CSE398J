```python
import pandas as pd 
import category_encoders as ce
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
heart=pd.read_csv('../input/heart-failure-prediction/heart.csv')
heart=pd.DataFrame(heart)
```

```python
heart.describe()
```

```python
sns.heatmap(heart.corr())
```

```python
train, test = train_test_split(heart,test_size=0.2,random_state= 1234)
```

```python
train['Sex'] = np.where(train['Sex'] == "F", 0, 1)
train['ExerciseAngina'] = np.where(train['ExerciseAngina'] == "N", 0, 1)
test['Sex'] = np.where(test['Sex'] == "F", 0, 1)
test['ExerciseAngina'] = np.where(test['ExerciseAngina'] == "N", 0, 1)
```

```python
train=pd.get_dummies(train)
test=pd.get_dummies(test)
```

```python
train.head()
```

```python
x_train=train.drop(['HeartDisease'],1)
x_test=test.drop(['HeartDisease'],1)

y_train=train['HeartDisease']
y_test=test['HeartDisease']
```

A BUNCH OF CLASSIFIERS USED 


```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

DECISION TREE CLASSIFICATION

```python
dct_clf=DecisionTreeClassifier()
model_dct=dct_clf.fit(x_train, y_train)
y_pred_dct=model_dct.predict(x_test)
accuracy_score(y_pred_dct,y_test)
```

RANDOM FOREST CLASSIFICATION

```python
rf_clf=RandomForestClassifier()
model_rf=rf_clf.fit(x_train, y_train)
y_pred_rf=model_rf.predict(x_test)
accuracy_score(y_pred_rf,y_test)
```

GRADIENT BOOST CLASSIFICATION

```python
gbc_clf=GradientBoostingClassifier()
model_gbc=gbc_clf.fit(x_train, y_train)
y_pred_gbc=model_gbc.predict(x_test)
accuracy_score(y_pred_gbc,y_test)
```
