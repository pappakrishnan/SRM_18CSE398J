## Digits Classification using Support Vector Machine

An SVC Model for classifying digits dataset.

### Gathering dataset and importing necessary modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
digits
```

```python
dir(digits)
```

```python
digits.target_names
```

### Cleaning of dataset

```python
df=pd.DataFrame(digits.data,digits.target)
df
```

```python
df['target']=digits.target
df
```

### Splitting dataset into train and test data for Classification

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('target',axis=1)
X
```

```python
y=df.target
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=40)
```

```python
len(X_train)
```

```python
len(X_test)
```

### SVC Classification

```python
from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
lin_model = SVC(kernel='linear')
lin_model.fit(X_train,y_train)
```

```python
lin_model.score(X_test,y_test)
```
