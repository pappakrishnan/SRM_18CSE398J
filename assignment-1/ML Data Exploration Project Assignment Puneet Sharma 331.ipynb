```python
ax = df['Price from the first transaction of a trading day'].hist(bins=30, grid=False, color='green', figsize=(8, 5)) # grid turned off and color changed

ax.set_xlabel('Price from the first transaction of a trading day')
ax.set_ylabel('Frequency')

ax.set_xlim(0, 1) # limiting display range to 0-70 for the x-axis
ax.set_ylim(0, 50); # limiting display range to 0-120 for the y-axis
```

```python
df_avg_BP = df.groupby('Price from the first transaction of a trading day')['Price from the last transaction of a trading day'].mean()
df_avg_BP[:5].plot.bar(color='orange');
```

```python
ax = df[['Price from the first transaction of a trading day', 'Price from the last transaction of a trading day']][:5].plot.bar(stacked=True)

ax.set_xlabel("Index")
ax.set_ylabel("Count");
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

df['Price from the first transaction of a trading day'].plot(ax = axes[0][0], style='.', color='red') # top left
df['Price from the last transaction of a trading day'].plot(ax = axes[0][1], style='.', color='blue') # top right

df['Maximum price in a trading day'].plot.hist(bins=30, ax = axes[1][0], color='black') # bottom left
df['Minimum price in a trading day'].plot.hist(bins=20, ax = axes[1][1], color='gray') # bottom right

axes[0][0].set_xlabel('index')
axes[0][1].set_xlabel('index')
axes[1][0].set_xlabel('Maximum price in a trading day')
axes[1][1].set_xlabel('Minimum price in a trading day')

axes[0][0].set_ylabel('Price from the first transaction of a trading day')
axes[0][1].set_ylabel('Price from the last transaction of a trading day')

axes[0][0].set_ylim(0, 1)
axes[0][1].set_ylim(0, 1)

axes[1][0].set_xlim(0, 1)
axes[1][1].set_xlim(0, 1)

fig.tight_layout()
```
