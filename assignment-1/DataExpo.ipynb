```python
ax = df['Price'].hist(bins=30, grid=False, color='red', figsize=(15, 6)) # grid turned off and color changed

ax.set_xlabel('Price')
ax.set_ylabel('sq.ft')

ax.set_xlim(0, 12000) # limiting display range to 0-70 for the x-axis
ax.set_ylim(0, 15000)
```

```python
df_avg = df.groupby('Floor')['Units'].mean()
df_avg[:10].plot.bar(color='orange');
```

```python
ax = df[['Units', 'Days_Till_Available']][:5].plot.bar(stacked=True)

ax.set_xlabel("Index")
ax.set_ylabel("Count");
```

```python
fig, axes = plt.subplots(2, 2, figsize=(25, 16))

df['Units'].plot(ax = axes[0][0], style='.', color='red') # top left
df['Floor'].plot(ax = axes[0][1], style='.', color='blue') # top right

df['Days_Till_Available'].plot.hist(bins=30, ax = axes[1][0], color='black') # bottom left
df['Estimated_Vacancy'].plot.hist(bins=20, ax = axes[1][1], color='gray') # bottom right

axes[0][0].set_xlabel('index')
axes[0][1].set_xlabel('index')
axes[1][0].set_xlabel('Days_Till_Available')
axes[1][1].set_xlabel('Estimated_Vacancy')

axes[0][0].set_ylabel('Units')
axes[0][1].set_ylabel('Floor')

axes[0][0].set_ylim(20, 800)
axes[0][1].set_ylim(0, 60)

axes[1][0].set_xlim(0, 120)
axes[1][1].set_xlim(0, 0.8)

fig.tight_layout()
```
