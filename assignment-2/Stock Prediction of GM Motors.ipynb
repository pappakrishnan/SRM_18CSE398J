```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics


data = pd.read_csv(r'C:\Users\asus\Downloads\GM_data.csv')
```

```python
data.head()
```

```python
data.plot.scatter('date', 'close', figsize=(8,5));
```

```python
x = data[['high', 'low', 'open', 'volume']].values
y = data['close'].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x_train, y_train)
```

```python
print(regressor.coef_)
```

```python
print(regressor.intercept_)
```

```python
predicted = regressor.predict(x_test)                           
```

```python
print(predicted)
```

```python
dframe = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted':predicted.flatten()})
```

```python
dframe.head(25)
```

```python
print("Mean Absolute Error: ", metrics.mean_absolute_error(y_test, predicted))
print("Mean Squared Error: ", metrics.mean_squared_error(y_test, predicted))
print("Root Mean Square Error: ", np.sqrt(metrics.mean_squared_error(y_test, predicted)))
```
