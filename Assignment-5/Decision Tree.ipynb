```python
inputs = df.drop('Drug', axis = 'columns')
target = df['Drug']
from sklearn.preprocessing import LabelEncoder
le_Age = LabelEncoder()
le_Sex = LabelEncoder()
le_BP = LabelEncoder()
le_Cholesterol = LabelEncoder()
le_Na_to_k = LabelEncoder()
```

```python
inputs['Age_n'] = le_Age.fit_transform(inputs['Age'])
inputs['Sex_n'] = le_Age.fit_transform(inputs['Sex'])
inputs['BP_n'] = le_Age.fit_transform(inputs['BP'])
inputs['Cholesterol_n'] = le_Age.fit_transform(inputs['Cholesterol'])
inputs['Na_to_K_n'] = le_Age.fit_transform(inputs['Na_to_K'])
inputs.head()
```

```python
inputs_n = inputs.drop(['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K'], axis = 'columns')
```

```python
inputs_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```
