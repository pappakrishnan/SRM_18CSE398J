```python
#pairplot
plt.figure(figsize=(4,4))
sns.pairplot(df,height=1.5)
plt.show()
```

```python
#pearson correlation
corr = df.corr(method='pearson').abs()
fig=plt.figure(figsize=(8,6))
sns.heatmap(corr,annot=True)
plt.title('Pearson Correlation')
plt.show()

print(df.corr()['Survived'].abs().sort_values())
```
