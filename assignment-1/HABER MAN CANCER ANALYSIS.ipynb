```python
status_1=iris.loc[iris["status"]==1]
status_2=iris.loc[iris["status"]==2]

plt.plot(status_1["age"], np.zeros_like(status_1["age"]),"o",label="status_1")
plt.plot(status_2["age"],np.zeros_like(status_2["age"]),"o",label="status_2")
plt.title("AGE")
plt.grid()
plt.legend()

plt.show()
```

# CDF

```python
# A]  plotting CDF
count , bin_edge = np.histogram(status_1["age"],bins=10,density=True)

pdf=count/sum(count)
print(pdf)
print(bin_edge)

cdf=np.cumsum(pdf)
plt.plot(bin_edge[1:],pdf,color="red")
plt.plot(bin_edge[1:],cdf,color="green")


count , bin_edge = np.histogram(status_2["age"],bins=10,density=True)

pdf=count/sum(count)
print(pdf)
print(bin_edge)

cdf=np.cumsum(pdf)
plt.plot(bin_edge[1:],pdf,color="blue")
plt.plot(bin_edge[1:],cdf,color="yellow")
plt.legend(["status_1_pdf","status_1_cdf","status_2_pdf","status_2_cdf"])
plt.grid()
plt.title("CDF for AGE")
plt.show()
```

1]. cdf curves are overlapped very much

2]. 20% of patients who survived(status_1) are of age below 40

3]. 20% of patients who do not survive(status_2) are of age approx. 45

```python
# B]
count , bin_edge = np.histogram(status_1["year"],bins=10,density=True)

pdf=count/sum(count)
print(pdf)
print(bin_edge)

cdf=np.cumsum(pdf)
plt.plot(bin_edge[1:],pdf,color="red")
plt.plot(bin_edge[1:],cdf,color="green")

count , bin_edge = np.histogram(status_2["year"],bins=10,density=True)

pdf=count/sum(count)
print(pdf)
print(bin_edge)

cdf=np.cumsum(pdf)
plt.plot(bin_edge[1:],pdf,color="blue")
plt.plot(bin_edge[1:],cdf,color="yellow")
plt.legend(["status_1_pdf","status_1_cdf","status_2_pdf","status_2_cdf"])
plt.title("CDF for YEAR")
plt.grid()
plt.show()

```

1]. CDF curves are highly overlapped , we can't distinguish bewtween status_1 and status_2

```python
#C]
count , bin_edge = np.histogram(status_1["nodes"],bins=10,density=True)

pdf=count/sum(count)
print(pdf)
print(bin_edge)

cdf=np.cumsum(pdf)
plt.plot(bin_edge[1:],pdf,color="red")
plt.plot(bin_edge[1:],cdf,color="green")

count , bin_edge = np.histogram(status_2["nodes"],bins=10,density=True)

pdf=count/sum(count)
print(pdf)
print(bin_edge)

cdf=np.cumsum(pdf)
plt.plot(bin_edge[1:],pdf,color="blue")
plt.plot(bin_edge[1:],cdf,color="yellow")
plt.grid()
plt.legend(["status_1_pdf","status_1_cdf","status_2_pdf","status_2_cdf"])
plt.title("CDF for NODES")
plt.show()
```

1]. though the curves are overlapped, we can see that,
more than 85% of patients who survived(status_1) have nodes below 10.
80% of patients who do not survive(status_2) are having nodes more than 10

we should carryout bivariate analysis further

# Statistics(Mean,Median,IQR,Quantile,MAD)

```python
print("MEDIANS of age of status_1 and status_2 :\n")
print(np.median(status_1["age"]))
print(np.median(status_2["age"]))

print("\nMEDIANS of year of status_1 and status_2 :\n")
print(np.median(status_1["year"]))
print(np.median(status_2["year"]))

print("\nMEDIANS of nodes of status_1 and status_2 :\n")
print(np.median(status_1["nodes"]))
print(np.median(status_2["nodes"]))
```

```python
print("quantiles of age of status 1 and status2 :\n")
print(np.percentile(status_1["age"],np.arange(0,100,25)))
print(np.percentile(status_2["age"],np.arange(0,100,25)))

print("\nquantiles of year of status 1 and status2 :\n")
print(np.percentile(status_1["year"],np.arange(0,100,25)))
print(np.percentile(status_2["year"],np.arange(0,100,25)))

print("\nquantiles of nodes of status 1 and status2 :\n")
print(np.percentile(status_1["nodes"],np.arange(0,100,25)))
print(np.percentile(status_2["nodes"],np.arange(0,100,25)))
```

1].From quantiles of age,
for age below 34 there is no patients of status_2(died)
2]. From quantiles of nodes,
75% of patients with status_1(survival) have nodes maximum of 3.

```python
from statsmodels import robust
print("mean absolute median of age status_1 and status_2 : ")
print(robust.mad(status_1["age"]))
print(robust.mad(status_2["age"]))

print("\nmean absolute median of year status_1 and status_2 : ")
print(robust.mad(status_1["year"]))
print(robust.mad(status_2["year"]))

print("\nmean absolute median of nodes status_1 and status_2 : ")
print(robust.mad(status_1["nodes"]))
print(robust.mad(status_2["nodes"]))
```

# JOINT ANALYSIS

From the contour plot for status_1(survived):

1]. More points are in the range of 1960-1962 and patients age approx 48-54.

From the contour plot for status_2(died):

1]. More points are at approx 1964 and patients age approx 50-51.

From the contour plot for status_1(survived):
